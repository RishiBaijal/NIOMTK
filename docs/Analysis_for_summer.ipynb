```python
tp = 0
tn = 0
fp = 0
fn = 0
for i in range(len(predicted)):
    if (predicted[i] == True and int(ground_truth[i] + 0.5) == 1):
        tp+=1
    elif (predicted[i] == True and int(ground_truth[i] + 0.5) == 0):
        fp +=1
    elif (predicted[i] == False and int(ground_truth[i] + 0.5) == 1):
        fn+=1
    else:
        tn+=1
print "TP = ", tp
print "TN = ", tn
print "FP = ", fp
print "FN = ", fn
print "Accuracy = ", (tp + tn + 0.0)/(tp + tn + fp + fn)
print "Precision = ", (tp + 0.0)/ (tp + fp)
print "Recall = ", (tp + 0.0)/(tp + fn)
    
```
